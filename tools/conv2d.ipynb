```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage


# 输入图像
img = np.array([
    np.linspace(1,255,5),
    np.linspace(1,255,5),
    np.linspace(1,255,5),
    np.linspace(1,255,5),
    np.linspace(1,255,5)
])

# 卷积核
kernel = np.array([[1, 0, -1],
                 [1, 0, -1],
                 [1, 0, -1]])
plt.imshow(img)
```

```python
# 卷积后的图像
conv_img = ndimage.convolve(img, kernel)

plt.imshow(conv_img)
```

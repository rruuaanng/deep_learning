```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image


# 输入图像
img = Image.open('test.png')

# 卷积核
kernel = np.array([[[0, 1, 0],
                    [1,-4, 1],
                    [0, 1, 0]],
                    [[0, 1, 0],
                    [1,-4, 1],
                    [0, 1, 0]],
                    [[0, 1, 0],
                    [1,-4, 1],
                    [0, 1, 0]]])
plt.imshow(img)
```

```python
# 卷积后的图像
conv_img = ndimage.convolve(img, kernel)

plt.imshow(conv_img)
```
